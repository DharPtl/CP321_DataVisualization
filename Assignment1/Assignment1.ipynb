Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets.
```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('data/fueltypesall1991-2023.csv')
print(df.columns)
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
Data_Cities = df[(df['Fuel Type']=="Regular Unleaded Gasoline")]
Data_Cities = df[(df['Date'] >= "2022")]
mean_price = Data_Cities.mean()
mean_price.plot.bar()

plt.title("Mean Retail Price of Regular Unleaded Gasoline in Cities ")
plt.xlabel("City")
plt.ylabel("Mean Price")
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
Data_Cities = df[(df['Fuel Type']=="Regular Unleaded Gasoline") & (df['Date'] >= "2022-01-01") & (df['Date'] <= "2022-12-31")]
Ott_City = Data_Cities["Ottawa"]

#plot histogram
plt.hist(Ott_City ,bins=10, color='purple', alpha=0.7, edgecolor='black')

#Add labels and title 
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Mean Retail Price of Regular Unleaded Gasoline in Ottawa")

plt.show()
```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
#Gather data for Ottawa from Q2.
Data_Cities = df[(df['Fuel Type']=="Regular Unleaded Gasoline") & (df['Date'] >= "2022-01-01") & (df['Date'] <= "2022-12-31")]
Ott_City = Data_Cities["Ottawa"]
# Plot the ECDF

def ecdf(data):
    length_data = len(data)
    x = np.sort(data)
    y = np.arange(1, length_data+1) / length_data
    return x, y

x, y = ecdf(Ott_City)

plt.plot(x,y, color='purple')


# Add labels and title to the plot
plt.xlabel("Price")
plt.ylabel("ECDF")
plt.title("ECDF of Regular Unleaded Gasoline price in Ottawa for 2022")

# Show the plot
plt.show()
```
